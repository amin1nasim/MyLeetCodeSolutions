```python
# Solution 1 (Recursive DFS)
# Time : O(n 2^n)
# Memory : O(n)

class Solution:
    def subsetsWithDup(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        result = set()

        def dfs(nums, res):
            if not nums:
                nonlocal result
                result.add(res)
                return
            temp, next_nums = nums[0], nums[1:]
            dfs(next_nums, res)
            dfs(next_nums, res + (temp,))
        dfs(nums, tuple())
        return list(result)
```

```python
# Solution 1 (Recursive DFS -- bit more optimized)
# Time : O(n 2^n)
# Memory : O(n) 

class Solution:
    def subsetsWithDup(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        result = []
        def dfs(idx, res):
            nonlocal nums
            if idx == len(nums):
                result.append(res[:])
                return
            res.append(nums[idx])
            dfs(idx + 1, res)
            res.pop()
            
            while idx + 1 < len(nums) and nums[idx] == nums[idx + 1]:
                idx += 1 
            
            dfs(idx + 1, res)
                
            
                
        dfs(0, [])
        return result
```
