```python
# (n*mlogm) Where n is number of strings and m is length of strings.
# This solution is valid since m <= 100 (m is small)
from collections import defaultdict
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        result_dict = defaultdict(list)
        for text in strs:
            result_dict[tuple(sorted(text))].append(text)
        
        return [result_dict[i] for i in result_dict]
```
