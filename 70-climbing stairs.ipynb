```python
# Solution 1
# Permutation solution
import math
class Solution:
    def climbStairs(self, n: int) -> int:
        result = 0
        numberOfTwos = n // 2
        numberOfOnes = n % 2
        while numberOfTwos >= 0:
            result += math.comb(numberOfTwos + numberOfOnes, numberOfTwos) # same as all permutaitons (2's + 1's)! / (2's! * 1's!)
            numberOfTwos -= 1
            numberOfOnes += 2
        
        return result
```

```python
# Solution 2 
# Top-down memoization
class Solution:
    def climbStairs(self, n: int) -> int:
        mem = {0:1, 1:1}    
        def dfs(n):
            nonlocal mem
            if n >= 0:
                if n not in mem:
                    mem[n] = dfs(n-1) + dfs(n-2)
                return mem[n]
            return 0

        return dfs(n)
```

```python
# Solution 3
# Bottom-up DP (neetcode solution)
class Solution:
    def climbStairs(self, n: int) -> int:
        one, two = 1, 1
        for i in range(n-1):
            one, two = one + two, one
        return one
```
