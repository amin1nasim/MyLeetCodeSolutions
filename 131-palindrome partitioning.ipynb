```python
# Recursive Solution (DFS)
# Time : O(n 2^n)
# Memory : O(n)

class Solution:
    def partition(self, s: str) -> List[List[str]]:
        result = []
        def dfs(s, res):
            if not s:
                result.append(res)
                return
            
            for i in range(len(s)):
                if s[:i+1:] == s[i::-1]:
                    dfs(s[i+1::], res + [s[:i+1:]])
            return

        dfs(s, [])
        return result
```
