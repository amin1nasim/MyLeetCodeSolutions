```python
# Solution 1
class Solution:
    def countBits(self, n: int) -> List[int]:
        if n == 0:
            return [0,]
        if n == 1:
            return [0,1]

        # Base cases 00 -> 0 and 01 - > 1
        result = [0] * (n + 1)
        result[0], result[1] = 0, 1

        idx = 2
        jumpBack = count = 2
        
        # F[x] = F[X - 2^floor(log_2(X))] + 1
        while idx < len(result):
            result[idx] = 1 + result[idx - jumpBack]
            count -= 1
            if count == 0:
                jumpBack = count = 2 * jumpBack
            idx += 1

        return result
```
