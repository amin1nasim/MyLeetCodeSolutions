```python
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        l = 0
        char_dict = {}
        maxf = 0
        result = 0
        for idx, r in enumerate(s):
            char_dict[r] = char_dict.get(r, 0) + 1
            maxf = max(maxf, char_dict[r])
            if idx - l + 1 - maxf > k:
                char_dict[s[l]] -= 1
                l += 1
            result = max(idx - l + 1,  result)

        return result
```
