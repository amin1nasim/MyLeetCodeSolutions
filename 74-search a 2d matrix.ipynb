```python
# Solution 1
# O(2 * log(n))
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        #  Binary Search on the rows
        rl = 0
        rr = len(matrix) - 1
        while rl <= rr:
            rm = (rl + rr) // 2
            if matrix[rm][0] > target:
                rr = rm - 1
            else:
                rl = rm + 1
        # Binary Search on the columns
        if rr >= 0:
            cl = 0
            cr = len(matrix[0]) - 1
            while cl <= cr:
                cm = (cl + cr) // 2
                if matrix[rr][cm] == target:
                    return True
                elif matrix[rr][cm] > target:
                    cr = cm - 1
                else:
                    cl = cm + 1
            
        return False
```
