```python
class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        result = []
        def recursive(n, o=0, c=0, s=''):
            if o == c == n:
                result.append(s)
            if o < n:
                recursive(n, o=o+1, c=c, s=s+'(')
            if c < o:
                recursive(n, o=o, c=c+1, s=s+')')
        
        recursive(n)
        return result
```
