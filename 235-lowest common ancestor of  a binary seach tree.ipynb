```python
# Solution 1 (When you forget that it is a binary 'search' tree)
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        result = None
        
        def dfs(root):
            if root is None:
                return False, False
            
            nonlocal result
            if result is None:
                lp, lq = dfs(root.left)
                if lp and lq:
                    return True, True

                rp, rq = dfs(root.right)
                if rp and rq:
                    return True, True

                np = (lp or rp) if root.val != p.val else True
                nq = (lq or rq) if root.val != q.val else True

                if np and nq:
                    result = root

                return np, nq
            
            return True, True

        dfs(root)  
        return result
```

```python
# Solution 2 (When I realized it is a binary SEARCH tree!)
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        while root:
            if (root.val - p.val) * (root.val - q.val) > 0:
                if root.val < p.val:
                    root = root.right
                else:
                    root = root.left
            else:
                return root 
```
