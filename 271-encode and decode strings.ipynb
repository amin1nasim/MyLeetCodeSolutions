```python
# Solution 1
class Solution:
    def encode(self, strs: List[str]) -> str:
        encoded = ""
        for word in strs:
            encoded += f"{len(word)}@" + word
        return encoded

    def decode(self, s: str) -> List[str]:
        result = []
        str_number = ""
        i = 0
        while i < len(s):
            if s[i] != '@':
                str_number += s[i]
                i += 1 

            else:
                number = int(str_number)
                str_number = ""
                result.append(s[i+1: i+number+1])
                i = i+number+1
        return result
```

```python
# Solution 2 (cleaner)
class Solution:
    def encode(self, strs: List[str]) -> str:
        # Concatenate each string with its length and a delimiter '@'
        return "".join(f"{len(word)}@{word}" for word in strs)

    def decode(self, s: str) -> List[str]:
        result = []
        i = 0
        
        while i < len(s):
            # Locate the next '@' to find the length of the word
            delimiter_index = s.find('@', i)
            length = int(s[i:delimiter_index])
            i = delimiter_index + 1  # Move past the '@'
            
            # Extract the word using the length
            result.append(s[i:i+length])
            i += length  # Move to the next encoded word
        
        return result
```
