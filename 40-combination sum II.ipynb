```python
# Recursive solution (DFS on sorted candidates)
# Time : O(n * 2^n)
# Memory: O(n)

class Solution:
    def combinationSum2(self, candidates: List[int], target: int) -> List[List[int]]:
        candidates.sort()
        result = []

        def dfs(idx, l):
            nonlocal target, candidates, result

            if sum(l) > target:
                return False

            elif sum(l) == target:
                result.append(l)
                return False
            
            i = idx
            while i < len(candidates):
                if not dfs(i+1, l + [candidates[i]]):
                    break
                i += 1
                while i < len(candidates) and  candidates[i] == candidates[i-1]:
                    i += 1
            
            return True

        dfs(idx=0, l=[])
        return res
```
