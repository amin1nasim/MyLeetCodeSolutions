```python
# Heap Solution
# Time: O(K * logN + N)
# Memory: O(N)

class Solution:
    def kClosest(self, points: List[List[int]], k: int) -> List[List[int]]:
        distances = [(x**2 + y**2, idx) for idx, (x,y) in enumerate(points)]
        heapq.heapify(distances)
        return [points[heapq.heappop(distances)[1]] for _ in range(k)]
```
