```python
from collections import deque
class Solution:
    def wordBreak(self, s: str, wordDict: List[str]) -> bool:
        starts = deque([0])
        seen = set()
        while starts:
            start_idx = starts.pop()
            for word in wordDict:
                end_idx = start_idx + len(word)
                if start_idx + len(word) <= len(s) and end_idx not in seen and word == s[start_idx:end_idx]:
                    seen.add(end_idx)
                    starts.appendleft(end_idx)
                    if end_idx == len(s):
                        return True

        return False
```
