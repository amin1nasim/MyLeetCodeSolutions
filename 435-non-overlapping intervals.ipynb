```python
class Solution:
    def eraseOverlapIntervals(self, intervals: List[List[int]]) -> int:
        # Sort by start time
        intervals.sort()

        end = intervals[0][1]
        removed = 0

        # Remove the tasks with higher end time
        for task in intervals[1:]:
            if task[0] < end:
                removed += 1
                end = min(end, task[1])
            else:
                end = task[1]
        return removed
```
