```python
class Solution:
    def canPartition(self, nums: List[int]) -> bool:
        # Each partition should create half of the total
        m = sum(nums) / 2
        if m != int(m):
            return False
        m = int(m)

        dp = [False] * (m + 1)
        dp[0] = True
        for c in nums:
            for p in range(m, c - 1, -1):
                if dp[p - c]:
                    dp[p] = True
        return dp[-1]
```
