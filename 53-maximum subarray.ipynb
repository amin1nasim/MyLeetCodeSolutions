```python
# Solution 1
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        def find_first_pos(start_idx, nums):
            for i in range(start_idx, len(nums)):
                if nums[i] <= 0:
                    i += 1
                else:
                    return i
            else:
                return -1
        
        
        i = find_first_pos(0, nums)
        if i == -1:
            return max(nums)

        temp_sum = result = 0
        while i != -1 and i < len(nums):
            if nums[i] >= 0:
                temp_sum += nums[i]
                i += 1
                
            else:
                result = max(temp_sum, result)
                if temp_sum + nums[i] >= 0:
                    temp_sum += nums[i]
                    i += 1 
                else:
                    temp_sum = 0
                    i = find_first_pos(i, nums)


        return max(result, temp_sum)
```

```python
# Solution 2 (cleaner)
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        # Start with the first element as the initial subarray sum
        current_sum = max_sum = nums[0]

        # Iterate through the rest of the array
        for num in nums[1:]:
            # Include the current element in the subarray or start a new subarray
            current_sum = max(num, current_sum + num)
            # Update max_sum if current_sum is larger
            max_sum = max(max_sum, current_sum)

        return max_sum
```
