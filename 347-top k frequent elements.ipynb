```python
# Solution 1
# O(n + klogn)
import heapq
from collections import defaultdict

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        counter = defaultdict(int)
        for n in nums:
            counter[n] += 1
        
        heap_counts = [(-counter[n], n) for n in counter]
        heapq.heapify(heap_counts)

        result = []
        for _ in range(k):
            result.append(heapq.heappop(heap_counts)[1])
        return result
```

```python
# Solution 2
# O(n + klogn)
from collections import Counter
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        heap = [(-c, item) for item, c in Counter(nums).items()]

        heapq.heapify(heap)

        sol = []

        for _ in range(k):
            sol.append(heapq.heappop(heap)[1])

        return sol
```
