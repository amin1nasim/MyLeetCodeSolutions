```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        m = len(nums)
        result = [1] * m

        for i in range(1, m):
            result[i] = result[i-1] * nums[i-1]
        
        temp = 1
        for i in range(m-2, -1, -1):
            temp *= nums[i+1]
            result[i] *= temp
            
        return result
```
