```python
# O(nlogn) because of sort
from collections import deque
class Solution:
    def isNStraightHand(self, hand: List[int], groupSize: int) -> bool:
        groupCount = len(hand) / groupSize
        if groupCount != int(groupCount):
            return False
        hand.sort()

        queue = deque()
        
        for card in hand:
            if queue:
                if queue[0][0] == card:
                    queue.append((card, 1))
                elif queue[0][0] + 1 == card:
                    _, rank = queue.popleft()
                    queue.append((card, rank + 1))
                else:
                    return False
            else:
                queue.append((card, 1))

            if queue[-1][1] == groupSize:
                queue.pop()
        return True if not queue else False
```
