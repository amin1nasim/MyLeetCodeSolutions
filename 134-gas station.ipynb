```python
class Solution:
    def canCompleteCircuit(self, gas: List[int], cost: List[int]) -> int:
        if sum(gas) < sum(cost):
            return -1

        n = len(gas)
        start_idx = 0
        curr_sum = 0

        for i in range(n):
            curr_sum += gas[i] - cost[i]
            if curr_sum < 0 :
                start_idx = i + 1
                curr_sum = 0
        return start_idx 
```
