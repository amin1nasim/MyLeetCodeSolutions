```python
# O(n)
from collections import deque
class Solution:
    def checkValidString(self, s: str) -> bool:
        queue = deque()
        stack = deque()

        for idx, char in enumerate(s):
            if char == '(':
                stack.append(idx)
            
            elif char == ')':
                if stack: # we have (
                    stack.pop()     
                elif queue: # we have *
                    queue.popleft()    
                else: # we have nothing :(
                    return False

            elif char == '*':
                queue.append(idx)

        while stack:
            if queue and queue[-1] > stack[-1]:
                queue.pop()
                stack.pop()
            else:
                return False
        return True
```
