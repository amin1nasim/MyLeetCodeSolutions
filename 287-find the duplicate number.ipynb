```python
class Solution:
    def findDuplicate(self, nums: List[int]) -> int:
        # Let's think of this question as a linked list question where for each element:
        # * Index of it (i) specifis its value
        # * Value of it (nums[i]) specifies the value of the next node.
        # We are interested in paths that have a rho shape (some nodes out of the cycle and then the cycle)
        # In this question it is easy to see the node with value 0 is always out of the loop (the next pointer value start with 1)
        # So we use Floyds algorithm to find the start of the cylce (repeated value)
        # Neetcode solution:
        slow, fast = 0, 0
        while True:
            slow = nums[slow]
            fast = nums[nums[fast]]
            if slow == fast:
                break

        slow2 = 0
        while True:
            slow = nums[slow]
            slow2 = nums[slow2]
            if slow == slow2:
                return slow
```
