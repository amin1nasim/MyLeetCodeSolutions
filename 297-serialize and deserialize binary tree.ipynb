```python
class Codec:

    def serialize(self, root):
        """Encodes a tree to a single string.
        
        :type root: TreeNode
        :rtype: str
        """    
        def dfs(root):
            if root:
                return str(root.val) + ', ' + dfs(root.left) + ', ' + dfs(root.right)
            else:
                return 'n'
        
        return dfs(root)

    def deserialize(self, data):
        """Decodes your encoded data to tree.
        
        :type data: str
        :rtype: TreeNode
        """
        data = data.split(', ')
        idx = 0

        def recursive_reconstruct():
            nonlocal idx, data
            if data[idx] != 'n':
                node = TreeNode(int(data[idx]))
                idx += 1
                node.left = recursive_reconstruct()
                node.right = recursive_reconstruct()
                return node
            else:
                idx += 1
                return None
        return recursive_reconstruct()
```
