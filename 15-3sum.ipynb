```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        result = []
        nums.sort()  # Sort the array to enable two-pointer technique
        
        for i in range(len(nums) - 2):  # Loop through the array
            if nums[i] > 0 :
                break
                
            if i > 0 and nums[i] == nums[i - 1]:  # Skip duplicate values for the first number
                continue

            # Two-pointer approach
            l, r = i + 1, len(nums) - 1
            while l < r:
                total = nums[i] + nums[l] + nums[r]
                
                if total < 0:  # If the sum is less than 0, move the left pointer right
                    l += 1
                elif total > 0:  # If the sum is greater than 0, move the right pointer left
                    r -= 1
                else:
                    result.append([nums[i], nums[l], nums[r]])  # Found a triplet
                    l += 1
                    r -= 1
                    # Skip duplicates for the second and third numbers
                    while l < r and nums[l] == nums[l - 1]:
                        l += 1
                    while l < r and nums[r] == nums[r + 1]:
                        r -= 1

        return result
```
