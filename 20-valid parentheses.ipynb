```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = deque()
        # Map closing brackets to their corresponding opening brackets
        bracket_map = {')': '(', ']': '[', '}': '{'}
        
        for char in s:
            if char in bracket_map.values():  # Opening brackets
                stack.append(char)
            else:  # Closing brackets
                if not stack or stack.pop() != bracket_map[char]:
                    return False
        
        return not stack
```
