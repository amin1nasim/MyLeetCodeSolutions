```python
# Solution 1 (Hashmap)
# Time: O(n) 
# Memory: O(n)

class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        seenNodes = set()

        while head:
            if head in seenNodes:
                return True
            seenNodes.add(head)
            head = head.next
        return False
```

```python
# Solution 2 (Fast & slow pointers)
# Time: O(n) 
# Memory: O(1)

class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        if not head:
            return False
        fasthead = head.next
        while fasthead is not None and fasthead.next is not None:
            if fasthead == head:
                return True
            head = head.next
            fasthead = fasthead.next.next
        return False
```
