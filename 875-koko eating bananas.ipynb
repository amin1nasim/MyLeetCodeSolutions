```python
from math import ceil
class Solution:
    def minEatingSpeed(self, piles: List[int], h: int) -> int:
        upper_bound = max(piles)
        lower_bound = 1
        while lower_bound <= upper_bound:
            mid = (upper_bound + lower_bound) // 2

            time_needed = 0
            for p in piles:
                time_needed += ceil(p/mid)
            
            if time_needed > h:
                lower_bound = mid + 1
            
            else:
                upper_bound = mid - 1
        return lower_bound
```
