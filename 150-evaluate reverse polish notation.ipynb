```python
class Solution:
    def evalRPN(self, tokens: List[str]) -> int:
        stack = deque()
        operations = '+-/*'
        for t in tokens:
            if t in operations:
                r = stack.pop()
                l = stack.pop()
                if t == '+':
                    stack.append(l+r)
                elif t == '-':
                    stack.append(l-r)
                elif t == '/':
                    stack.append(int(l/r))
                elif t == '*':
                    stack.append(l*r)
            else:
                stack.append(int(t))
        return stack[0]
```
