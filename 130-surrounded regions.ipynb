```python
class Solution:
    def solve(self, board: List[List[str]]) -> None:
        """
        Do not return anything, modify board in-place instead.
        """
        m, n = len(board), len(board[0])
        bfs = deque()
        borderIndices = [(i, j) for i in [0, m-1] for j in range(n)] +\
                        [(i, j) for i in range(1, m-1) for j in [0, n-1]]
        boundaryCheck = lambda i, j: 0 <= i < m and 0 <= j < n
        directions = [(1, 0), (0, 1), (-1, 0), (0, -1)]

        # Find the "O"s on border
        for i, j in borderIndices:
            if board[i][j] == 'O':
                bfs.append((i, j))
        
        # Change regions with border "O" to "B".
        while bfs:
            for _ in range(len(bfs)):
                i,j = bfs.popleft()
                board[i][j] = 'B'
                for ii, jj in directions:
                    newI, newJ = i+ii, j+jj
                    if boundaryCheck(newI, newJ) and board[newI][newJ] == 'O':
                        bfs.append((newI, newJ))
        
        # Change "B" to "O" and everything else to "X"
        for i in range(m):
            for j in range(n):
                if board[i][j] == 'B':
                    board[i][j] = 'O'
                else:
                    board[i][j] = 'X'
        
        return board
```
