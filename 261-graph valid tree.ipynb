```python
from collections import defaultdict
class Solution:
    def validTree(self, n: int, edges: List[List[int]]) -> bool:
        adjMap = defaultdict(list)
        for v0, v1 in edges:
            adjMap[v0].append(v1)
            adjMap[v1].append(v0)
        
        seen = set()
        def dfs(node, prev = -1):
            nonlocal seen, adjMap
            if node in seen:
                return False
            seen.add(node)
            for e in adjMap[node]:
                if e != prev:
                    if dfs(e, node) == False:
                        return False
            return True
        
        return dfs(0) and len(seen) == n # No loop and have seen all the vertices
```
