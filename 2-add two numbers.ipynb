```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
```

```python
# Solution 1 (Itration)
# Time: O(n+m) where n and m are lengths of l1 and l2
# Memory: O(1) (It isn't implace but I don't consider result's memory)

class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        carry = 0
        head = dummy = ListNode()

        # Add the two lists
        while l1 and l2:
            val = l1.val + l2.val  + carry
            if val >= 10:
                carry = 1
                head.next = ListNode(val = val - 10)
            else:
                carry = 0
                head.next = ListNode(val = val)
            head = head.next
            l1 = l1.next
            l2 = l2.next

        # Add the elementes of the longer list
        longerList = l1 or l2
        while longerList:
            val = longerList.val + carry
            if val >= 10:
                carry = 1
                head.next = ListNode(val = val - 10)
            else:
                carry = 0
                head.next = ListNode(val = val)
            head = head.next
            longerList = longerList.next

        # At the end if carry is not 0 add it
        if carry:
            head.next = ListNode(val=1)

        return dummy.next
```

```python
# Solution 2 (Recursive)
# Time: O(n+m) where n and m are lengths of l1 and l2
# Memory: O(n+m)

class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        def recursiveSum(l1, l2, c):
            if l1 or l2 or c:
                newNode = ListNode()
                val = (l1.val if l1 else 0) + (l2.val if l2 else 0) + c
                if val >= 10:
                    newNode.val = val - 10
                    c = 1
                else:
                    newNode.val = val
                    c = 0
                newNode.next = recursiveSum(l1.next if l1 else None, l2.next if l2 else None, c)
                return newNode

            else:
                return None
            
        return recursiveSum(l1, l2, 0)
```
