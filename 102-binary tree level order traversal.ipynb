```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
```

```python
# BFS Solution
# Time: O(n)
# Memory: O(2^d) where d is the depth of the tree

from collections import deque
class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        bfs = deque()
        result = []
        if root:
            bfs.appendleft(root)
            while bfs:
                layer = []
                for _ in range(len(bfs)):
                    top = bfs.pop()
                    layer.append(top.val)
                    if top.left: bfs.appendleft(top.left)
                    if top.right: bfs.appendleft(top.right)
                result.append(layer)
        return result
```
