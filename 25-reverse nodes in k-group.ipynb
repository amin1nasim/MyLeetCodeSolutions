```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
```

```python
# Iterative Solution
# Time: O(n^2)
# Memory: O(1)

class Solution:
    def reverseKGroup(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        if not head:
            return None

        def reverse(head, k):
            prev = None
            curr = head
            for _ in range(k):
                temp = curr.next
                curr.next = prev
                prev, curr = curr, temp
            return prev, head

        prevNode = None
        start = pointer = head
        while pointer:
            for _ in range(k):
                if pointer:
                    pointer = pointer.next
                else:
                    break
            else:
                start, end = reverse(start, k)
                end.next = pointer
                 
                if prevNode:
                    prevNode.next = start
                else:
                    head = start
                prevNode = end
                start = pointer
        
        return head
```
