```python
# Solution 1 (Recursive DFS)
# Time: O(n * 2^n) -> We create 2^n nodes and for each node we copy the list wiht n elements
# Memory: O(n)

class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        result = []
        
        def dfs(nums, curr=None):
            if nums:
                curr = curr if curr is not None else []
                dfs(nums[1:], curr)
                dfs(nums[1:], curr + [nums[0]])
            
            else:
                nonlocal result
                result.append(curr)

        dfs(nums)
        return result
```

```python
# Solution 2 (Iterative BFS)
# Time: O(n * 2^n)
# Memory: O(n) -> we have to keep nums and this algorithm is not inplace

class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        queue = deque([[]])
        for i in nums:
            for _ in range(len(queue)):
                sublist = queue.popleft()
                queue.append(sublist)
                queue.append(sublist + [i])
        return list(queue)
```

```python
# Solution 3 (Bit Manipulation)
# Time: O(n * 2^n)
# Memory: O(n)

class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        total_subsets = 2 ** len(nums)
        result = []
        for i in range(total_subsets):
            result.append([n for j, n in enumerate(nums) if (i >> j) & 0b1])
        return result
```
