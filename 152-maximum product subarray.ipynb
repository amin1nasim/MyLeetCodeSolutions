```python
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        result = submin = submax = nums[0]

        for i in nums[1:]:
            if i < 0:
                tempMax, tempMin = submin * i, submax * i
            else:
                tempMax, tempMin = submax * i, submin * i

            if tempMax > submax:
                submax = tempMax
            else:
                result = max(result, submax)
                submax = max(i, tempMax)
            
            submin = tempMin if tempMin < submin else min(i, tempMin)

        return max(result, submax)
```
