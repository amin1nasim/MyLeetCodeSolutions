```python

# O(n) where n is size of s.
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s)!=len(t):
            return False
            
        word_count = {}
        for char in s:
            if char in word_count:
                word_count[char] += 1
            else:
                word_count[char] = 1
        
        for char in t:
            if char not in word_count:
                return False
            word_count[char] -= 1    
            if word_count[char] == 0:
                del(word_count[char])
        
        return True if len(word_count)  == 0 else False
```

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False

        count = [0] * 26
        for i in range(len(s)):
            count[ord(s[i]) - ord('a')] += 1
            count[ord(t[i]) - ord('a')] -= 1

        for val in count:
            if val != 0:
                return False
        return True
```
