```python
# Solution 1 (sorting)
# O(N logN)
class Solution:
    def canAttendMeetings(self, intervals: List[Interval]) -> bool:
        if not intervals:
            return True

        intervals.sort(key=lambda x: x.start)
        
        for idx in range(1, len(intervals)):
            prev_task, next_task = intervals[idx - 1], intervals[idx]
            if prev_task.end > next_task.start:
                return False
            prev_task = next_task
        return True 
```

```python
# Solution 2
# O(N * M) where M is max element in the list
class Solution:
    def canAttendMeetings(self, intervals: List[Interval]) -> bool:
        if not intervals:
            return True

        latest_end = max(intervals, key=lambda x:x.end).end
        occupied = [False] * (1 + latest_end)

        for task in intervals:
            for idx in range(task.start, task.end):
                if occupied[idx]:
                    return False
                occupied[idx] = True
        return True
```
