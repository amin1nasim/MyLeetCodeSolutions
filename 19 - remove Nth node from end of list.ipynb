```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
```

```python
# Solution 1 (Recursive)
# Time: O(n)
# Space: O(n)

class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        def recursive (curr, n):
            if curr is None:
                return None

            curr.next = recursive(curr.next, n)
            n[0] -= 1
            if n[0] == 0:
                return curr.next
            return curr

        return recursive(head, [n])
```

```python
# Solution 2 (Iteration)
# Time: O(n)
# Space: O(1)

class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        # Count to total number of nodes
        count = 1
        temp = head
        while temp.next:
            temp = temp.next
            count += 1
            
        # Calculate which node should be deleted from the start
        deleteIdx = count - n + 1
        
        # Delete the node 
        if deleteIdx == 1:
            return head.next
        
        temp = head
        for _ in range(deleteIdx - 2):
            temp = temp.next
        temp.next= temp.next.next
        return head
```

```python
# Solution 3 (Two Pointers)
# Time: O(n)
# Space: O(1)

class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        p1, p2 = head, head
        for _ in range(n):
            p2 = p2.next
            
        if p2 is None:
            return p1.next
        
        while p2.next:
            p2 = p2.next
            p1 = p1.next
        
        p1.next = p1.next.next
        return head
```
