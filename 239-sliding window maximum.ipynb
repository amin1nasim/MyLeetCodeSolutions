```python
# Solution 1
# Heap O(nlogn)
class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        heap = [(-v, i) for i, v in enumerate(nums[:k])]
        heapify(heap)
        result = [-heap[0][0]]

        for i, v in enumerate(nums[k:], start=k):
            heappush(heap, (-v, i))
            while i - heap[0][1] >= k:
                heappop(heap)
            result.append(-heap[0][0])
        return result
```

```python
# Solution 2
# Dequeu O(n)
class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        dq = deque()
        result = []

        for i, v in enumerate(nums[:k]):
            while dq and dq[-1][1] <= v:
                dq.pop()
            dq.append((i, v))

        result.append(dq[0][1])

        for i, v in enumerate(nums[k:], start=k):
            while dq and dq[-1][1] <= v:
                dq.pop()
            dq.append((i, v))
            if i - dq[0][0] >= k:
                dq.popleft()
            result.append(dq[0][1])
        return result 
```
