```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
```

```python
# Solution 1
# Makes a copy of the orginal lists
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        # If one of the linkedlists are empty
        if list1 is None:
            return list2
        if list2 is None:
            return list1
        
        # Find the new head
        if list1.val < list2.val:
            head = ListNode(val=list1.val)
            list1 = list1.next
        else:
            head = ListNode(val=list2.val)
            list2 = list2.next
        
        # Add the new nodes to new linked list
        lastNode = head
        while list1 is not None or list2 is not None:
            v1 = list1.val if list1 is not None else float('inf')
            v2 = list2.val if list2 is not None else float('inf')
            if v1 < v2:
                newNode = ListNode(val=v1)
                list1 = list1.next

            else:
                newNode = ListNode(val=v2)
                list2 = list2.next
            lastNode.next = newNode
            lastNode = lastNode.next
        return head
```

```python
# Solution 2
# Inplace merge
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode()
        last = dummy
        while list1 and list2:
            if list1.val < list2.val:
                last.next = list1
                list1 = list1.next
            else:
                last.next = list2
                list2 = list2.next
            last = last.next
        if list1:
            last.next = list1
        if list2:
            last.next = list2
        return dummy.next
```
