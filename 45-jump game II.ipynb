```python
# Solution 1 (Dynamic Programming)
# O(n^2)
class Solution:
    def jump(self, nums: List[int]) -> int:
        n = len(nums)
        dp = [n] * n
        dp[-1] = 0

        for idx in range(n-2, -1, -1):
            for j in range(idx + 1, idx + nums[idx] + 1):
                if j < n:
                    dp[idx] = min(dp[idx], dp[j] + 1)
        return dp[0]
```

```python
# Solution 2 (Greedy)
# O(n)
class Solution:
    def jump(self, nums: List[int]) -> int:
        n = len(nums)
        if n == 1:
            return 0
        
        currentMaxReach = nextMaxReach = nums[0]
        steps = i = 1
        while i <= currentMaxReach and currentMaxReach < n - 1:
            nextMaxReach = max(i + nums[i], nextMaxReach)

            if i == currentMaxReach:
                steps += 1
                currentMaxReach = nextMaxReach
            i += 1
        return steps
```
