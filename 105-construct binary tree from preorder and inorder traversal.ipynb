```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
```

```python
# Solution 1 (Recursive - DFS)
# Time: o(n^2)
# Memory: O(n) (worse case)

class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
        def recursive(preorder_idx, inorder):
            if not inorder:
                return None

            nonlocal preorder
            node = TreeNode(val=preorder[preorder_idx])
            idx = inorder.index(node.val)
            node.left = recursive(preorder_idx + 1, inorder[:idx])
            node.right = recursive(preorder_idx + idx + 1, inorder[idx+1:])
            return node
        
        return recursive(preorder_idx=0, inorder=inorder)
```

```python
# Solution 2 (Recursive - DFS + Hashmap)
# Time: o(n)
# Memory: O(n) (worse case)

class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
        inorder_val2index = {val:idx for idx, val in enumerate(inorder)}
        preorder_idx = 0

        def recursive(l, r):
            if l>r:
                return None

            nonlocal preorder, preorder_idx, inorder_val2index
            node = TreeNode(val=preorder[preorder_idx])
            preorder_idx += 1
            idx = inorder_val2index[node.val]
            node.left = recursive(l, idx - 1)
            node.right = recursive(idx + 1, r)
            return node
        
        return recursive(0, len(inorder)-1)
```
