```python
# Solution 1 (Heap)
# Time: O(M) where M is number of tasks
# Memory: O(26) Alphabet size

class Solution:
    def leastInterval(self, tasks: List[str], n: int) -> int:
        time = 0
        tasks = [-v for v in Counter(tasks).values()]
        heapq.heapify(tasks)
        waitList = deque()

        while tasks or waitList:
            time += 1
            if waitList and time - waitList[0][1] > n:
                heapq.heappush(tasks, waitList.popleft()[0])

            if tasks:
                freq = heapq.heappop(tasks) + 1
                if freq !=0:
                    waitList.append((freq, time))
        
        return time
```

```python
# Solution 2 (Math)
# Time: O(M)
# Memory: O(26)

class Solution:
    def leastInterval(self, tasks: List[str], n: int) -> int:
        counter = collections.Counter(tasks)
        max_count = max(counter.values())
        min_time = (max_count - 1) * (n + 1) + \
                    sum(map(lambda count: count == max_count, counter.values()))
    
        return max(min_time, len(tasks))
```
