```python
# Solution 1 (New Tree)
# Time: O(n)
class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        def dfs(root):
            if root is None:
                return None
            
            new_root = TreeNode(root.val, left=dfs(root.right), right=dfs(root.left))
            return new_root
        return dfs(root)
```

```python
# Solution 2 (Inplace)
# Time: O(n)
class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        def dfs(root):
            if root is None:
                return None
            root.left, root.right = root.right, root.left
            dfs(root.left)
            dfs(root.right)

        dfs(root)
        return root
```
