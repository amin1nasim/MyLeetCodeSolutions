```python
from collections import defaultdict
class Solution:
    def isValidSudoku(self, board: List[List[str]]) -> bool:
        row_set = [set() for i in range(9)]
        column_set = [set() for i in range(9)]
        box_set = [set() for i in range(9)]
        

        for r in range(9):
            for c in range(9):
                b = r // 3 * 3 + c // 3
                val = board[r][c]
                if val != '.':
                    if (val in row_set[r]) or (val in column_set[c]) or (val in box_set[b]):
                        return False
                    
                    row_set[r].add(val)
                    column_set[c].add(val)
                    box_set[b].add(val)
        return True
```
