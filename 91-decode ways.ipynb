```python
class Solution:
    def numDecodings(self, s: str) -> int:
        if not s or s[0] == '0':  # Early return for invalid cases
            return 0

        # Possible decoding of the substring from size 0 to length of s
        possibleDecodings = [0] * (len(s) + 1)

        # Base case: empty string has one way to decode
        possibleDecodings[0] = 1

        # Base case: string of size one (not '0') has one way to decode 
        possibleDecodings[1] = 1
        
        for idx in range(1, len(s)):
            # Check single digit decode
            single = 1 if s[idx] != '0' else 0 
            # Check two-digit decode
            double = 1 if 10 <= int(s[idx - 1: idx + 1]) <= 26 else 0
            possibleDecodings[idx+1] = single * possibleDecodings[idx] + double * possibleDecodings[idx - 1]

            if possibleDecodings[idx+1] == 0:
                return 0
        return possibleDecodings[-1]
```
