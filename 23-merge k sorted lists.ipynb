```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
```

```python
# Heap Solution
# Time: O(n logk)
# Memory: O(k)
# Where n is the total number of nodes and k is the number of linklists.
import heapq
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        heap = [(node.val, idx) for idx, node in enumerate(lists) if node is not None]
        heapq.heapify(heap)

        dummy = pointer = ListNode() 
        while heap:
            _, idx = heapq.heappop(heap)
            pointer.next = lists[idx]
            pointer = pointer.next
            if lists[idx].next:
                heapq.heappush(heap, (lists[idx].next.val, idx))
                lists[idx] = lists[idx].next
        return dummy.next
```
