```python
# Solution 1 (more complete implementation)
# O(n)
class Solution:
    def isPalindrome(self, s: str) -> bool:
        s = s.lower()
        l, r = 0, len(s) - 1

        while l < r:
            if not s[l].isalnum():
                l += 1
                
            elif not s[r].isalnum():
                r -= 1
            
            elif s[l] == s[r]:
                r -= 1
                l += 1

            else:
                return False
            
        return True
```

```python
# Solution 2 (using pythonic hacks)
# O(n)
class Solution:
    def isPalindrome(self, s: str) -> bool:
        x = ''
        for c in s:
            if c.isalnum():
                x += c.lower()
        return x == x[::-1]
```
