```python
class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        total = len(nums1) + len(nums2)
        halfTotal = total // 2
        smaller, bigger = (nums1, nums2) if len(nums1) < len(nums2) else (nums2, nums1)
        
        l = 0
        r = len(smaller) - 1
        
        while True:
            ils = (l + r) // 2
            ilb = halfTotal - ils - 2
            vls = smaller[ils] if ils > -1 else float("-inf") 
            vrs = smaller[ils + 1] if ils + 1 < len(smaller) else float("inf")
            vlb = bigger[ilb] if ilb > -1 else float("-inf") 
            vrb = bigger[ilb + 1] if ilb + 1 < len(bigger) else float("inf")
            
            if vls <= vrb and vlb <= vrs:
                if total % 2 == 0:
                    return (max(vls, vlb) + min(vrs, vrb)) / 2
                else:
                    return min(vrs, vrb)

            if vls > vrb:
                r = ils - 1
            else:
                l = ils + 1
```
