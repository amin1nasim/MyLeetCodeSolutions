```python
class Solution:
    def rob(self, nums: List[int]) -> int:
        first = nums[0]
        if len(nums) == 1:
            return first

        second = max(nums[0], nums[1])
        if len(nums) == 2:
            return second
        
        for i in range(2, len(nums)):
            third = max(first + nums[i], second)
            first, second = second, third
        
        return third
```
