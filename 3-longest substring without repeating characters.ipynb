```python
# O(n)
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        r = l = 0
        char_set = set()
        result = 0
        while r < len(s):
            if s[r] not in char_set:
                char_set.add(s[r])
            else:
                result = max(result, r - l)
                while s[l] != s[r]:
                    char_set.remove(s[l])
                    l += 1
                l += 1
            r += 1
        return max(result, r - l)
```
