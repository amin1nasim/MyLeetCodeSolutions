```python
class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        n = len(cost)
        oneStep = 0
        twoStep = 0

        for step in range(n-3, -2, -1):
            twoStep, oneStep = (oneStep,
                                min(cost[step + 1] + oneStep, cost[step + 2] + twoStep))
        return oneStep
```
