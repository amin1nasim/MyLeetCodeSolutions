```python
# Solution 1 O(n^2)
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        dp = [1] * len(nums)

        for i in range(1, len(nums)):
            for j in range(i):
                if nums[i] > nums[j]:
                    dp[i] = max(dp[i], 1 + dp[j])
        return max(dp)
```

```python
# Solution 2
#  O(N logN)
#  https://leetcode.com/problems/longest-increasing-subsequence/solutions/1326308/c-python-dp-binary-search-bit-segment-tree-solutions-picture-explain-o-nlogn
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        def bSearch(l, num):
            s = 0
            e = len(l) - 1
            while s <= e:
                mid = (s+e) // 2

                if l[mid] >= num:
                    e = mid - 1
                else:
                    s = mid + 1
            return s

        result = [nums[0],]
        for num in nums:
            if num > result[-1]:
                result.append(num)
            elif num == result[-1]:
                continue
            else:
                result[bSearch(result, num)] = num
        return len(result)
```
