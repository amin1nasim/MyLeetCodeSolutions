```python
class Solution:
    def mergeTriplets(self, triplets: List[List[int]], target: List[int]) -> bool:
        x = y = z = False

        for triple in triplets:
            if triple[0] > target[0] or triple[1] > target[1] or triple[2] > target[2]:
                continue
            x |= (triple[0] == target[0])
            y |= (triple[1] == target[1])
            z |= (triple[2] == target[2])
            if x and y and z:
                return True
        
        return False
```
