```python
class Solution:
    def carFleet(self, target: int, position: List[int], speed: List[int]) -> int:
        n = len(position)
        sorted_idx = [i for i in range(n)]
        sorted_idx.sort(reverse=True, key=lambda x: position[x])

        def eta(p, s):
            nonlocal target
            return (target - p) / s

        stack = deque([eta(position[sorted_idx[0]], speed[sorted_idx[0]])])
        for i in sorted_idx[1:]:
            cur_eta = eta(position[i], speed[i])
            if cur_eta > stack[-1]:
                stack.append(cur_eta)
        
        return len(stack)
```
