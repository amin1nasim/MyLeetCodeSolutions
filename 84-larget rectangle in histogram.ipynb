```python
class Solution:
    def largestRectangleArea(self, heights: List[int]) -> int:
        stack = deque()
        result = 0
        for i, h in enumerate(heights):
            start = i
            while stack and stack[-1][1] >= h:
                j, hj = stack.pop()
                result = max(result, (i - j) * hj)
                start = j
            stack.append((start, h))
        i += 1
        while stack:
            j, hj = stack.pop()
            result = max(result, (i - j) * hj)
        return result
```
