```python
# The challenge is to not sort the array, but still ...
# Solution 1 (Sorting)
# Time: O(N log N)
# Memory: O(N) or O(1) depending on the sorting algorithm

class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        nums.sort()
        return nums[len(nums)-k]
```

```python
# Solution 2 (Heap)
# Time: O(K logN + N) -> Pops element from a heap size B. If K~N then it becomes the same time complexity as sorting.
# Memory: O(1)

class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        k = len(nums) - k + 1

        heapq.heapify(nums)
        for _ in range(k):
            result = heapq.heappop(nums)
        
        return result
```

```python
# Solution 3 (Heap of size K)
# Time: O(N log K) -> Pushes all the elements into a heap of size K. If K~N then it becomes the same time complexity as sorting.
# Memory: O(K) -> This one creates a seperate array unlike heapify.

class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        return heapq.nlargest(k, nums)[-1]
```

```python
# Solution 3 (Quick Sort)
# Time: Average Case O(N), Worse Case O(N^2)
# Memory: Depends on the number of recusion calls -> Best Case: O(log N) Worse Case: O(N)

class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        k = len(nums) - k

        def quickSort(l, r):
            pivot, pointer = nums[r], l
            for i in range(l, r):
                if nums[i] < pivot:
                    nums[i], nums[pointer] = nums[pointer], nums[i]
                    pointer += 1
            nums[pointer], nums[r] = nums[r], nums[pointer]
            
            if pointer < k:
                return quickSort(pointer + 1, r)
            
            elif pointer > k:
                return quickSort(l, pointer - 1)
            
            else:
                return pivot
        
        return quickSort(0, len(nums)-1)
```
