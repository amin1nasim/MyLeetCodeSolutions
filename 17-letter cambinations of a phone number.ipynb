```python
# BFS Solution just to do it differently
# Time : O(4^(n+1))
# Memory: O(n)

class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        if len(digits) == 0:
            return []

        numCharMap = {'2':'abc', '3':'def', '4':'ghi', '5':'jkl', '6':'mno', '7':'pqrs', '8':'tuv', '9':'wxyz'}
        queue = deque([""])
        for d in digits:
            for _ in range(len(queue)):
                temp = queue.popleft()
                for c in numCharMap[d]:
                    queue.append(temp + c)

        return list(queue)
```
