```python
# O(n)
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        set_nums = set(nums)
        
        maximum = 0
        while set_nums:
            temp = 1
            i = start = set_nums.pop()

            while i + 1 in set_nums:
                set_nums.remove(i + 1)
                i += 1
                temp += 1
            
            i = start
            while i - 1 in set_nums:
                set_nums.remove(i - 1)
                i -= 1
                temp += 1
            
            maximum = max(temp, maximum)
        
        return maximum
```
