```python
class Solution:
    def reverse(self, x: int) -> int:
        INT_MAX_9DIGIT = 214748364  # MAX:  2147483647
        INT_MIN_9DIGIT = -214748364 # MIN: -2147483648
        result = 0

        while x != 0:
            r = x % 10 if x > 0 else x % -10
            x = int(x/10)

            if result > INT_MAX_9DIGIT  or (result == INT_MAX_9DIGIT  and r > 7):
                return 0
            if result < INT_MIN_9DIGIT  or (result == INT_MIN_9DIGIT  and r < -8):
                return 0
            
            result = result * 10 + r
        
        return result

```
