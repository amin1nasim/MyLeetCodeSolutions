```python
# Recursive Solution (DFS)
# Time : O(n! n^2)
# Memory : O(n! n)
class Solution:
    def permute(self, nums: List[int]) -> List[List[int]]:
        result = []

        def dfs(nums, curr=[]):
            if not nums:
                result.append(curr)
            
            for i in range(len(nums)):
                nxt = curr + [nums[i]]
                dfs(nums[:i] + nums[i+1:], nxt)
        dfs(nums)
        return result
```
