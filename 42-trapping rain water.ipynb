```python
# O(n)
class Solution:
    def trap(self, height: List[int]) -> int:
        n = len(height)
        if n < 3:
            return 0
        result = 0
        l = 0
        r = n - 1
        maxl = height[l]
        maxr = height[r]

        while l < r:
            if maxl <= maxr:
                l += 1
                if height[l] < maxl:
                    result += maxl - height[l]
                else:
                    maxl = height[l]
            else:
                r -= 1
                if height[r] < maxr:
                    result += maxr - height[r]
                else:
                    maxr = height[r]
        return result
```
