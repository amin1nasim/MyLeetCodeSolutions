```python
class Solution:
    def longestPalindrome(self, s: str) -> str:
        start = end = 0

        def expand(idx1, idx2):
            while idx1 - 1 >= 0 and idx2 + 1 < len(s) and s[idx1 - 1] == s[idx2 + 1]:
                idx1 -= 1
                idx2 += 1

            return idx1, idx2
                

        for idx, _ in enumerate(s):
            # Odd size palindrom
            if (idx - 1 >= 0) and (idx + 1 < len(s)) and (s[idx - 1] == s[idx + 1]):
                temp_s, temp_e = expand(idx - 1, idx + 1)
                if temp_e - temp_s > end - start:
                    start, end = temp_s, temp_e
            # Even size palindrom
            if idx + 1 < len(s) and s[idx] == s[idx + 1]:
                temp_s, temp_e = expand(idx, idx + 1)
                if temp_e - temp_s > end - start:
                    start, end = temp_s, temp_e
            
            
        return s[start: end + 1]
```
