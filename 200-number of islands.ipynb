```python
# Recuersive Solution (DFS)
# Time: O(MN) Wher the board is MXN
# Memory: O(MN)

class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        m, n = len(grid), len(grid[0])
        directions = [(1, 0), (0, 1), (-1, 0), (0, -1)]
        result = 0

        def inside(i, j):
            if 0 <= i < m and 0 <= j < n:
                return True
            return False

        def dfs(i, j):
            nonlocal grid, directions
            grid[i][j] = "2"
            for di, dj in directions:
                if inside(i + di, j + dj) and grid[i+di][j+dj] == "1":
                    dfs(i + di, j + dj)

        for i in range(m):
            for j in range(n):
                if grid[i][j] == "1":
                    result += 1
                    dfs(i, j)
        
        return result
```
