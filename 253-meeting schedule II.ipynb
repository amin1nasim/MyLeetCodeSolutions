```python
"""
Definition of Interval:
class Interval(object):
    def __init__(self, start, end):
        self.start = start
        self.end = end
"""

class Solution:
    def minMeetingRooms(self, intervals: List[Interval]) -> int:
        if not intervals:
            return 0

        intervals.sort(key = lambda x: x.start)
        days = [intervals[0].end, ]

        for task in intervals[1:]:
            for day_index, end_time in enumerate(days):
                if end_time <= task.start:
                    days[day_index] = task.end
                    break
            else:
                days.append(task.end)
        return len(days)
```
