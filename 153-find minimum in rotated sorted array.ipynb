```python
# O(log(n))
class Solution:
    def findMin(self, nums: List[int]) -> int:
        # If nums[0] < nums[-1]
        ## Then the array is sorted
        # If nums[0] == nums[-1]
        ## Then the array is size 1 as the numbers are unique
        if nums[0] <= nums[-1]:
            return nums[0]
        l = 1
        r = len(nums) - 1

        while l <= r:
            mid = (l + r) // 2
            if nums[mid] > nums[0]:
                l = mid + 1
            elif nums[mid] < nums[0]:
                r = mid - 1
            # The only cases that mid is 0
            ## l = r = 0
            ## l = 0 r = 1 (from first if we know that len(num)>1 )
            elif nums[mid] == nums[0]:
                return min(nums[0], nums[1])
        
        return nums[l]
```
