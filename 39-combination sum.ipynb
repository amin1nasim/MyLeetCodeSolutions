```python
# Recursive Solution (DFS on sorted candidates)
# Time : O(2^(t/m) * (t/m)) (Check later) The largest depth of dfs will be target (t) divided by
#                     minimum element in candidates (m). Lets say each element can either say or can be remove
#                     be replace by another element so 2^(t/m). Additionally each time we create a new list of number so
#                     a (t/m) will be the upper bound on the size of copy that recusion performs.
# Memory: O(2^(t/m))

class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        candidates.sort()
        result = []

        def dfs(idx, l):
            nonlocal candidates, target
            if sum(l) > target:
                return False

            elif sum(l) == target:
                result.append(l)
                return False
            
            else:
                for i, num in enumerate(candidates[idx:], start=idx):
                    if not dfs(i, l + [num]):
                        break
                return True

        dfs(idx=0, l=[])
        return result
```
