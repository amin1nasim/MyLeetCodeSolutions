```python
# Solution 1 (Dynamic Programming)
#  O(n^2)
class Solution:
    def canJump(self, nums: List[int]) -> bool:
        n = len(nums)
        isPossible = [False] * n
        isPossible[-1] = True

        for i in range(n - 2, -1, -1):
            for j in range(i+1, i+nums[i]+1):
                if j < n:
                    if isPossible[j]:
                        isPossible[i] =  True
                        break
                else:
                    break

        return isPossible[0]
```

```python
# Solution 2 (Greedy)
# O(n)
class Solution:
    def canJump(self, nums: List[int]) -> bool:
        n = len(nums)
        if n == 1: return True
        
        maxReach = nums[0]
        i = 1
        while i <= maxReach and maxReach < n-1:
            maxReach = max(i + nums[i], maxReach)
            i += 1

        return True if maxReach >= n - 1 else False
```
