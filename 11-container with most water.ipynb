```python
# O(n)
class Solution:
    def maxArea(self, height: List[int]) -> int:
        result = 0
        l = 0 
        r = len(height) - 1
        while l < r:
            if height[l] > height[r]:
                result = max(result, height[r] * (r - l))
                r -= 1
            elif height[l] < height[r]:
                result = max(result, height[l] * (r - l))
                l += 1
            else:
                result = max(result, height[l] * (r - l))
                l += 1
                r -= 1
        return result
```
