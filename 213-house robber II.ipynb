```python
class Solution:
    def rob(self, nums: List[int]) -> int:
        ## f the value we rob if there were one house -
        # excluding the last house
        ## g the value we rob if there were one house -
        # excluding the house number 1
        f1, g1 = 0, 0
        if len(nums) == 1:
            return nums[0]

        f2, g2 = nums[0], nums[1]
        if len(nums) == 2:
            return max(nums[0], nums[1])
        
        for i in range(2, len(nums)):    
            f3 = max(f1 + nums[i-1], f2)
            g3 = max(g1 + nums[i], g2)
            f1, g1, f2, g2 = f2, g2, f3, g3
        return max(f3, g3) 
```
