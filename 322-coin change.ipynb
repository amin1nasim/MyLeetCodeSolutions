```python
class Solution:
    def coinChange(self, coins: List[int], amount: int) -> int:
        subProb = [float('inf')] * (amount + 1)
        subProb[0] = 0  # Base case: 0 coins are needed to make amount 0

        for i in range(1, amount + 1):
            for coin in coins:
                if i - coin >= 0:
                    subProb[i] = min(subProb[i], 1 + subProb[i - coin])

        return subProb[amount] if subProb[amount] != float('inf') else -1
```
