```python
    def kthSmallest(self, root: Optional[TreeNode], k: int) -> int:
        result = None
        def dfs(root, k):
            if root is None:
                return 0
            
            seen = dfs(root.left, k) + 1
            if seen > k:
                return seen
            elif seen == k:
                nonlocal result
                result = root.val
                return seen
            else:
                return dfs(root.right, k-seen) + seen
            
        dfs(root, k)
        return result
```
